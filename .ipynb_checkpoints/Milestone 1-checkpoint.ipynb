```python
import nibabel as nib
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
nibimgs = []
nibhdrs = []
filenames = []
for file,subdir,d in os.walk(os.getcwd()):
    if 'analyze' in file:
        filenames.append(file)
        a =[x for x in d if 'img' in x][0]
        imgpath = (os.path.join(file,a))
        nibimgs.append(imgpath)
        b = [x for x in d if 'hdr' in x][0]
        hdrpath = os.path.join(file,b)
        nibhdrs.append(hdrpath)
        
```

```python
datums = []
for file in nibimgs:
    print ('\n==================================================================\nProcessing file: ',os.path.split(file)[1])
    img = nib.load(file)
    hdr = img.header
    print ('Pixel Spacing')
    print (hdr.get_zooms()[0:21])
    print ('Slice thickness')
    print (hdr.get_zooms()[2])
    data = img.get_fdata()
    x,y,z = data.squeeze().shape
    middle_coronal = np.rot90(data[:,:,z//2],3)
    print ('Shape of output image below:')
    print (middle_coronal.squeeze().shape)
    plt.figure(figsize=(8,8))
    plt.title('Middle coronal slice %s' %(os.path.split(file)[1]))
    plt.imshow(middle_coronal.squeeze(),cmap='coolwarm')
    plt.show()
```
